La función `type()` nos dice de qué tipo es algún elemento en concreto. Veamos!

```python
# Tipos de elementos
print(type(10))
print(type(3.14))
print(type(-8))
print(type(1e3))
print(type("30"))
print(type(1 + 3j))
```
